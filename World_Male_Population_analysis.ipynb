## INTRODUCTION

The dataset contains historical data on the male population worldwide from the year 2004 to 2024. It includes yearly population estimates, providing insights into demographic trends over two decades. This dataset is useful for analyzing population growth patterns,

## Data Preprocessing

Importing necessary libraries for the analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel('/content/World malepopulation 2024-2004.xlsx')
data.head(5)
```

*The data also giving the Counry names and country codes. Here the data in the year 2024 is empty, so we can remove that column. Also Series code and series name is also can be remove*

## Data Visualization

### Top 10 Highest Population Countries

```python
countries = ['India', 'China']
filtered_data = data[data['Country Name'].isin(countries)]
filtered_data.set_index('Country Name', inplace=True)

year_columns = [col for col in filtered_data.columns if col.startswith('YR') or 'YR' in col]
cleaned_years = []
for col in year_columns:
    if col.startswith('YR'):
        cleaned_years.append(col[2:])
    else:
        cleaned_years.append(col.split(' ')[0])

plot_data = filtered_data[year_columns].T
plot_data.index = cleaned_years
plot_data.index = plot_data.index.astype(int)

plt.figure(figsize=(12, 4))
for country in countries:
    plt.plot(plot_data.index, plot_data[country], label=country)

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population of India, China (2004-2023)')
plt.legend()
plt.xticks([2004, 2023])
plt.show()
```
